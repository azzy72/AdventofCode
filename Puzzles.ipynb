```python
### Part 1 ###

with open("./InputFolder/input_day2.txt", "r") as infile:
    reports = []
    for line in infile:
        reports.append(line.strip().split(" "))


def TryAscending(input_report : list):
    """
    test if a list is ascending (increasing : return True) or descending (decreasing : return False)
    """
    previous_level = int(input_report[0])
    level_diff = []
    for level in input_report[1:]:
        level_diff.append(int(level) - previous_level)
        previous_level = int(level)
    
    return level_diff


def Evaluate_leveldiff(level_diff : list, debug : bool = False):
    up_count = 0
    down_count = 0
    na_count = 0
    if debug: print("\n",level_diff)

    for res in level_diff:
        if 1 <= res <= 3 : up_count += 1
        elif -3 <= res <= -1: down_count += 1
        else: na_count += 1
    
    return [up_count, down_count, na_count]


def SafeUnsafe(reports, debug : bool = False):
    Safe_count = 0
    Unsafe_count = 0

    for report in reports:
        cEvaluation = Evaluate_leveldiff(TryAscending(report), debug=debug)
        if cEvaluation[2] > 0: 
            if debug: print(f"Unsafe, due to wrong change:\t {cEvaluation}")
            Unsafe_count += 1
        elif cEvaluation[0] > 0 and cEvaluation[1] > 0: 
            if debug: print(f"Unsafe, increasing & decreasing: {cEvaluation}")
            Unsafe_count += 1
        else: 
            if debug: print(f"Safe:\t\t\t\t {cEvaluation}")
            Safe_count += 1

    return [Safe_count, Unsafe_count]

print(f"Safe vs Unsafe: {SafeUnsafe(reports, debug=False)}")
```

```python
### Part 2 ###

with open("./input_day2.txt", "r") as infile:
    reports = []
    for line in infile:
        reports.append(line.strip().split(" "))


def TryAscending(input_report : list):
    """
    Test if a list is ascending
    """
    previous_level = int(input_report[0])
    level_diff = []
    for level in input_report[1:]:
        level_diff.append(int(level) - previous_level)
        previous_level = int(level)
    
    return level_diff


def Evaluate_leveldiff(level_diff : list, debug : bool = False):
    up_count = 0
    down_count = 0
    na_count = 0
    if debug: print("\n",level_diff)

    for res in level_diff:
        if 1 <= res <= 3 : up_count += 1
        elif -3 <= res <= -1: down_count += 1
        else: na_count += 1
    
    return [up_count, down_count, na_count]

def SafeUnsafe_Nice(rep, debug : bool = False):
    cEvaluation = Evaluate_leveldiff(TryAscending(rep), debug=debug)

    if cEvaluation[2] > 1: 
        if debug: print(f"Unsafe, due to wrong change:\t {cEvaluation}")
        return False
    elif cEvaluation[0] > 1 and cEvaluation[1] > 1: 
        if debug: print(f"Unsafe, increasing & decreasing: {cEvaluation}")
        return False
    elif 0 not in cEvaluation:
        if debug: print(f"Unsafe no zero count:\t\t {cEvaluation}")
        return False
    else: 
        if debug: print(f"Safe:\t\t\t\t {cEvaluation}")
        return True

def Removal_nice(report):
    for i in range(len(report)):
        cReport = report.copy()
        del cReport[i]
        print(cReport)
        if SafeUnsafe_Nice(cReport):
            print("niced",cReport)
            return True


Safe_count = 0 
Unsafe_count = 0 
for report in reports:  
    if SafeUnsafe_Nice(report):
        Safe_count += 1
    elif Removal_nice(report):
        Safe_count += 1
    else:
        Unsafe_count += 1

print(f"Safe {Safe_count} vs Unsafe: {Unsafe_count}")
```

```python
import re
with open("./InputFolder/input_day3.txt", "r") as infile:
    file = "".join(infile.readlines()) #file as one string

def ExtractMul(mul_sec):
    mul_reg = r"mul\([\d]{1,3},[\d]{1,3}\)" #regex pattern for mul(123,123)
    res_re = re.findall(mul_reg, mul_sec)

    res_mul = 0
    for mul in res_re:
        numbers = mul[4:-1].split(",")
        res_mul += int(numbers[0]) * int(numbers[1])

    return res_mul


def FindTrueMul(file, debug : bool = False):
    TotalMul = 0
    sec_split = file.split("don't")
    TotalMul += ExtractMul(sec_split[0])
    for sec_neg in sec_split[1:]:
        if debug: print("--- New Iteration ---")
        if "do" in sec_neg:
            if debug: print(sec_neg)
            sec_pos = sec_neg.split("do")[1:]
            for pos in sec_pos:
                if debug: print(pos)
                TotalMul += ExtractMul(pos)
            if debug: print("")

    return TotalMul

print(FindTrueMul(file))
```

```python
import numpy as np

testDataset_in = [["M", "M", "M", "S", "X", "X", "M", "A", "S", "M"], 
                  ["M", "S", "A", "M", "X", "M", "S", "M", "S", "A"],
                  ["A", "M", "X", "S", "X", "M", "A", "A", "M", "M"],
                  ["M", "S", "A", "M", "A", "S", "M", "S", "M", "X"],
                  ["X", "M", "A", "S", "A", "M", "X", "A", "M", "M"],
                  ["X", "X", "A", "M", "M", "X", "X", "A", "M", "A"],
                  ["S", "M", "S", "M", "S", "A", "S", "X", "S", "S"],
                  ["S", "A", "X", "A", "M", "A", "S", "A", "A", "A"],
                  ["M", "A", "M", "M", "M", "X", "M", "M", "M", "M"],
                  ["M", "X", "M", "X", "A", "X", "M", "A", "S", "X"]]

collectionoflines = []
with open("./InputFolder/input_day4.txt", "r") as infile:
    for line in infile:
        cCollect = []
        for letter in line.strip():
            cCollect.append(letter)
        collectionoflines.append(cCollect) #list in list

#collectionoflines = testDataset_in #toggle for test dataset

class CeresSearch:
    """
    **horizontalSearch()**: performs search horizontally, both left to right and right to left (by flipping order)
    **verticalSearch()**: performs search vertically
    **diagonalSearch()**: performs search diagonally
    #####
    **self.data** <list(list(str))> : is initialized when creating class *(required : given)*
    **self.data_transposed** <np.array> : transposed input data in np.array structure
    **self.searchword** <str> : word to search for *(required : given)*
    **self.searchword_size** <int> : length of search word *(calculated : not given)*
    """

    def __init__(self, data, searchword : str):
        self.data = np.array(data)
        #self.data_transposed = np.transpose(self.data)
        self.searchword = searchword
        self.searchword_size = len(searchword)
        

    def horizontalSearch(self, inv_pos, row_pos, debug_mode : bool = False):
        """
        Search for xmas horizontally, right and left\n
        **inv_pos** is the position of an X in a list (row)\n
        **row_pos** is the iter number of the current row (in self.data)
        """
        right = ""
        left = ""
        failed_orientations = [False, False]
        for i in range(self.searchword_size):
            #Orientation right
            if failed_orientations[0] == False and len(self.data[row_pos]) - inv_pos >= 3:
                try:
                    right += str(self.data[row_pos][inv_pos+i])
                except IndexError as err:
                    if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
                if debug_mode: print(f"hsearch\titer_layer: {i}\t[right]", right)
            
            #Orientation left
            if failed_orientations[1] == False and inv_pos >= 3:
                try:
                    left += str(self.data[row_pos][inv_pos-i])
                except IndexError as err:
                    if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
                if debug_mode: print(f"hsearch\titer_layer: {i}\t[left]", left)

            #Raising flags if wrongful spelling in each orientation
            if right[0:i+1] != self.searchword[0:i+1]: failed_orientations[0] = True
            if left[0:i+1] != self.searchword[0:i+1]: failed_orientations[1] = True

            #If none starts with searchword; Early termination
            if False not in failed_orientations:
                if debug_mode: print(f"All orientations failed\nRight: {right}\nLeft: {left}")
                if debug_mode: print(f"Iteration: {i}, current flags: {failed_orientations}\n")
                return None
        
            if debug_mode: print(f"Iteration: {i}, current flags: {failed_orientations}\n")

        return [right, left]


    def verticalSearch(self, inv_pos, row_pos, debug_mode : bool = False):
        """
        Search for xmas vertically, up and down\n
        Transposed data structure makes horizontal search vertical\n
        *inv_pos* is the position of an X in a list (row)\n
        *row_pos* is the iter number of the current row (in self.data)
        """
        up = ""
        down = ""
        failed_orientations = [False, False]
        for i in range(self.searchword_size):
            #Orientation up
            if failed_orientations[0] == False and row_pos >= 3:
                try:
                    up += str(self.data[row_pos-i][inv_pos])
                except IndexError as err:
                    if debug_mode: print(f"index of size {len(self.data)} out of bounds.\nError: {err}")
                if debug_mode: print(f"vsearch\titer_layer: {i}\t[top]", up)
            
            #Orientation down
            if failed_orientations[1] == False and len(self.data) - row_pos >= 3:
                try:
                    down += str(self.data[row_pos+i][inv_pos])
                except IndexError as err:
                    if debug_mode: print(f"index of size {len(self.data)} out of bounds.\nError: {err}")
                if debug_mode: print(f"vsearch\titer_layer: {i}\t[bottom]", down)

            #Raising flags if wrongful spelling in each orientation
            if up[0:i+1] != self.searchword[0:i+1]: failed_orientations[0] = True
            if down[0:i+1] != self.searchword[0:i+1]: failed_orientations[1] = True

            #If none starts with searchword; Early termination
            if False not in failed_orientations:
                if debug_mode: print(f"All orientations failed\nUp: {up}\nDown: {down}")
                if debug_mode: print(f"Iteration: {i}, current flags: {failed_orientations}\n")
                return None
        
            if debug_mode: print(f"Iteration: {i}, current flags: {failed_orientations}\n")

        return [up, down]


    def diagonalSearch(self, inv_pos, row_pos, debug_mode : bool = False):
        """
        Search for xmas diagonally, up and down, left and right\n
        *inv_pos* is the position of an X in a list (row)\n
        *row_pos* is the iter number of the current row (in self.data)
        """
        topright = ""
        topleft = ""
        bottomright = ""
        bottomleft = ""
        failed_orientations = [False, False, False, False]
        for i in range(self.searchword_size):        
            #Orientation: top right
            if failed_orientations[0] == False and row_pos >= 3 and len(self.data[row_pos]) - inv_pos >= 3: 
                try: 
                    topright += str(self.data[row_pos-i][inv_pos+i])
                except IndexError as err:
                    if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
                if debug_mode: print(f"dsearch\titer_layer: {i}\t[top right]", topright)

            #Orientation: top left
            if failed_orientations[1] == False and row_pos >= 3 and inv_pos >= 3: #no need to run wrongfully spelling orientations
                try: 
                    topleft += str(self.data[row_pos-i][inv_pos-i])
                except IndexError as err:
                    if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
                if debug_mode: print(f"dsearch\titer_layer: {i}\t[top left]", topleft)

            #Orientation: bottom right
            if failed_orientations[2] == False and len(self.data) - row_pos >= 3 and len(self.data[row_pos]) - inv_pos >= 3: #no need to run wrongfully spelling orientations
                try: 
                    bottomright += str(self.data[row_pos+i][inv_pos+i])
                except IndexError as err:
                    if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
                if debug_mode: print(f"dsearch\titer_layer: {i}\t[bottom right]", bottomright)

            #Orientation: bottom left
            if failed_orientations[3] == False and len(self.data) - row_pos >= 3 and inv_pos >= 3: #no need to run wrongfully spelling orientations
                try:
                    bottomleft += str(self.data[row_pos+i][inv_pos-i])
                except IndexError as err:
                    if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
                if debug_mode: print(f"dsearch\titer_layer: {i}\t[bottom left]", bottomleft)

            #Raising flags if wrongful spelling in each orientation
            if topright[0:i+1] != self.searchword[0:i+1]: failed_orientations[0] = True
            if topleft[0:i+1] != self.searchword[0:i+1]: failed_orientations[1] = True
            if bottomright[0:i+1] != self.searchword[0:i+1]: failed_orientations[2] = True
            if bottomleft[0:i+1] != self.searchword[0:i+1]: failed_orientations[3] = True

            #If none starts with searchword; Early termination
            if False not in failed_orientations:
                if debug_mode: print(f"All orientations failed\nTop Right: {topright}\nTop Left: {topleft}\nBottom Right: {bottomright}\nBottom Left: {bottomleft}")
                if debug_mode: print(f"Iteration: {i}, current flags: {failed_orientations}\n")
                return None
        
            if debug_mode: print(f"Iteration: {i}, current flags: {failed_orientations}\n")

        return [topright, topleft, bottomright, bottomleft]
    

    def search(self, inv_pos, data):
        """
        Perform search in a list both left to right or right to left
        """
        test = data
        hoz_test = "".join(test[inv_pos:inv_pos+self.searchword_size])
        if hoz_test == self.searchword and len(test) - inv_pos >= 3: return hoz_test

        #reverse order horizontal - right to left
        test_rev = self.reverselist(test)
        hozrev_test = "".join(test_rev[inv_pos:inv_pos+self.searchword_size])
        if hozrev_test == self.searchword and inv_pos > 2: return hozrev_test
    

    def reverselist(self, list):
        """
        Reverse order in list
        """
        new_list = []
        for ent in reversed(list):
            new_list.append(ent)
        return new_list

    def XMASdiagonal(self, inv_pos, row_pos, debug_mode : bool = False):
        """
        Checks diagonal positions when A positions (row_pos, inv_pos) is found 
        """
        topright = ""
        topleft = ""
        bottomright = ""
        bottomleft = ""
        FoundMAS = [False, False]
        #Orientation: top right
        if row_pos >= 1 and len(self.data[row_pos]) - inv_pos >= 1: 
            try: 
                topright = str(self.data[row_pos-1][inv_pos+1])
            except IndexError as err:
                if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
            if debug_mode: print(f"dsearch\t[top right]", topright)

        #Orientation: top left
        if row_pos >= 1 and inv_pos >= 1: #no need to run wrongfully spelling orientations
            try: 
                topleft = str(self.data[row_pos-1][inv_pos-1])
            except IndexError as err:
                if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
            if debug_mode: print(f"dsearch\t[top left]", topleft)

        #Orientation: bottom right
        if len(self.data) - row_pos >= 1 and len(self.data[row_pos]) - inv_pos >= 1: #no need to run wrongfully spelling orientations
            try: 
                bottomright = str(self.data[row_pos+1][inv_pos+1])
            except IndexError as err:
                if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
            if debug_mode: print(f"dsearch\t[bottom right]", bottomright)

        #Orientation: bottom left
        if len(self.data) - row_pos >= 1 and inv_pos >= 1: #no need to run wrongfully spelling orientations
            try:
                bottomleft = str(self.data[row_pos+1][inv_pos-1])
            except IndexError as err:
                if debug_mode: print(f"index of size {len(self.data[row_pos])} out of bounds.\nError: {err}")
            if debug_mode: print(f"dsearch\t[bottom left]", bottomleft)

        #Raising flags if wrongful spelling in each orientation
        if topright+"A"+bottomleft == self.searchword or bottomleft+"A"+topright == self.searchword: FoundMAS[0] = True
        if topleft+"A"+bottomright == self.searchword or bottomright+"A"+topleft == self.searchword: FoundMAS[1] = True
    
        #If none starts with searchword; Early termination
        if False in FoundMAS:
            if debug_mode: print(f"No Found MAS\nTopRight-BottomLeft: {topright}A{bottomleft}\nBottomLeft-TopRight: {bottomleft}A{topright}\nTopLeft-BottomRight: {topleft}A{bottomright}\nBottomRight-TopLeft: {bottomright}A{topleft}")
            if debug_mode: print(f"Current flags: {FoundMAS}\n")
            return None
    
        if debug_mode: print(f"Current flags: {FoundMAS}\n")

        return FoundMAS




```

```python
def LoadFiles(Test : bool = False, debug_mode : bool = False):
    """
    Load files for day 5. Returns list where [0] = Rules_in <dict>, and [1] = Updates_in <list>
    *Test* <bool>: if True, load smaller example file, else load input file
    *debug_mode* <bool>: if True, execute prints for troubleshooting / debugging
    
    Thoughts on Rules_in structure:\n
    Rules (eg. 97|13 & 97|47) are extracted in lists of length = 2; (eg. [97,13] & [97,47]).
    The latter number is stored as keys in Rules_in, and the former as a value <list> (eg. Rules_in[97] = [13, 47]).
    As such, if a subsequent page number in an update list, is found as a value of its preceeding page number in Rules_in; The update fails
    """
    Rules_in = dict()
    Updates_in = []
    divider_flag = False
    if Test: Path_to_file = "./ExamplesFolder/example_day5.txt" #example file
    else: Path_to_file = "./InputFolder/input_day5.txt" #input file

    with open(Path_to_file, "r") as infile:
        for line in infile:
            if line == "\n": #divides section 1 and 2
                divider_flag = True
                continue
            
            if divider_flag: #Marks section 2
                cUpdate = list(map(int, line.strip().split(","))) #Convert elements to integers using map()
                Updates_in.append(cUpdate) #One update is a list of numbers
            
            else: #Marks section 1
                cRule = list(map(int, line.strip().split("|"))) #Convert elements to integers using map()
                if cRule[0] in Rules_in.keys():
                    cValues = Rules_in[cRule[0]] #extract value list of already added values to the key: cRule[0]
                    cValues.append(cRule[1]) #add new value to value list
                    Rules_in[cRule[0]] = cValues #rewrite value list for key cRule[0]
                else: 
                    Rules_in[cRule[0]] = [cRule[1]]

    #Extraction successful as integers
    if debug_mode: print("Test - two integers in Updates_in summed:", Updates_in[0][1]+Updates_in[0][2])
    if debug_mode: print(f"Test - a key ({47}) has the following values (rules) in Rules_in:", Rules_in[47])

    return [Rules_in, Updates_in]

def reverselist(list):
    """
    Reverse order in list
    """
    new_list = []
    for ent in reversed(list):
        new_list.append(ent)
    return new_list
```

```python
Rules = LoadFiles(Test=False)[0] #dictionary of rules
Updates = LoadFiles(Test=False)[1] #list of list of updates
TS = True #Toggle troubleshooting in below code
Approved_Updates = 0 #counter for accepted updates
MiddleUpdateSum = 0 #sum for all middle numbers of accepted updates

for update in Updates:
    if TS: print(f"##### New update ##### - {update}")
    update_status = True #if False; failed update
    rev_update = reverselist(update) #reversed update. If preceeding numbers are found in rules; the update fails
    for i in range(len(rev_update)): #going through entries "factorial"-like (one less each iteration)
        #Control: need to break out of two loops in case an update has failed
        if update_status == False: 
            break
        
        #Control: end of update, continue to next
        if len(rev_update[i+1:]) < 1:
            if update_status: 
                cMiddleNumber = update[len(update) // 2] #Finding middle number
                if TS: print(f"  >Middle number recognized: {cMiddleNumber}")
                MiddleUpdateSum += cMiddleNumber
                if TS: print(f"Update correct: {update}\n")
                Approved_Updates += 1
            continue
        
        #Evaluation: evaluating the page numbers in updates, in relation to given rules
        if TS: print(f"Testing: {rev_update[i]}, remaining: {rev_update[i+1:]}")
        for preceeding in rev_update[i+1:]: #remaining page numbers in current update
            try: #Test for presence of rule for iter number in update.
                Rules[rev_update[i]]
            except KeyError as err: #If not, print then continue 
                if TS: print(f"  !No rules for {rev_update[i]}")
                break
            
            #Evaluating rules aligned with iter number
            if preceeding in Rules[rev_update[i]]:
                if TS: print("Update failed: ",update)
                if TS: print(f"  !Rules for failure of {rev_update[i]}: {Rules[update[i]]}\n")
                update_status = False
                break

print(f"Success rate of updates {Approved_Updates}/{len(Updates)}:\t\t\t{Approved_Updates/len(Updates)*100:.2f}%")
print(f"Sum of middle number (accepted updates):\t{MiddleUpdateSum}")
```
